```python
import requests

API_BASE_URL = "https://samples.openweathermap.org/data/2.5/forecast/hourly?q=London,us&appid=b6907d289e10d714a6e88b30761fae22"

def get_weather_forecast(date):
    url = API_BASE_URL
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        for forecast in data["list"]:
            if forecast["dt_txt"].startswith(date):
                temperature = forecast["main"]["temp"] - 273.15  # Convert Kelvin to Celsius
                return temperature
        return None
    else:
        return None

def get_wind_speed(date):
    url = API_BASE_URL
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        for forecast in data["list"]:
            if forecast["dt_txt"].startswith(date):
                wind_speed = forecast["wind"]["speed"]
                return wind_speed
        return None
    else:
        return None

def get_pressure(date):
    url = API_BASE_URL
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        for forecast in data["list"]:
            if forecast["dt_txt"].startswith(date):
                pressure = forecast["main"]["pressure"]
                return pressure
        return None
    else:
        return None

def main():
    while True:
        print("\nOptions:")
        print("1. Get weather")
        print("2. Get Wind Speed")
        print("3. Get Pressure")
        print("0. Exit")

        choice = int(input("Enter your choice: "))

        if choice == 0:
            print("Terminating the program.")
            break
        elif choice in (1, 2, 3):
            date = input("Enter the date (yyyy-mm-dd): ")

            if choice == 1:
                temperature = get_weather_forecast(date)
                if temperature is not None:
                    print(f"Temperature on {date}: {temperature:.2f}°C")
                else:
                    print("Failed to get the temperature.")
            elif choice == 2:
                wind_speed = get_wind_speed(date)
                if wind_speed is not None:
                    print(f"Wind Speed on {date}: {wind_speed} km/h")
                else:
                    print("Failed to get the wind speed.")
            elif choice == 3:
                pressure = get_pressure(date)
                if pressure is not None:
                    print(f"Pressure on {date}: {pressure} hPa")
                else:
                    print("Failed to get the pressure.")
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
